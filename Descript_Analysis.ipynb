```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load your synthetic dataset
df = pd.read_csv("hr_synthetic_dataset.csv")  
```

```python
df.head()
```

```python
sns.countplot(data=df, x='Gender') # Count plot for gender 
plt.title("Gender Distribution")
plt.show()
sns.countplot(data=df, x='MaritalStatus') # Count plot for marital status 
plt.title("Marital Status Distribution")
```

```python
sns.boxplot(data=df, x='Gender', y='MonthlyIncome') # Box plot for monthly income by gender 
plt.title("Monthly Income Distribution by Gender")
plt.show()
```

```python
promotion_rate = df[df['YearsSinceLastPromotion'] == 0].groupby('Gender').size() / df.groupby('Gender').size() # Calculate promotion rate for each gender   
promotion_rate.plot(kind='bar', title='Recent Promotion Rate by Gender') # Plot the promotion rate for each gender
plt.ylabel("Promotion Rate") 
plt.show()
```

```python
pd.crosstab(df['Gender'], df['JobLevel'], normalize='index').plot(kind='bar', stacked=True) # Stacked bar plot for job level distribution
plt.title("Job Level Distribution by Gender") 
plt.ylabel("Proportion")
plt.show()
```

```python
attrition_rate = df[df['Attrition'] == 'Yes'].groupby('Gender').size() / df.groupby('Gender').size() # Calculate attrition rate for each
attrition_rate.plot(kind='bar', title='Attrition Rate by Gender') # Plot attrition rate by gender
plt.ylabel("Attrition Rate")
plt.show()
```

```python
avg_salary_by_role_gender = df.groupby(['JobRole', 'Gender'])['MonthlyIncome'].mean().unstack() # Calculate average salary by job role
avg_salary_by_role_gender.plot(kind='bar', figsize=(12, 6)) # Plot average salary by job role 
plt.title("Avg Monthly Income by Job Role & Gender")
plt.ylabel("Salary")
plt.show()
```

```python
df['TimeToPromotion'] = df['YearsSinceLastPromotion']
df.groupby('Gender')['TimeToPromotion'].mean().plot(kind='bar', title="Avg Years Since Last Promotion") # Plot average years since last promotion
plt.ylabel("Years")
plt.show()
```
